# 2-4 연습문제
다음의 함수 $f(x)$에 대한 그래프를 그리시오.<br>
$f(x) = x^3-3x^2+4$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x ** 3 - 3*x**2 +4
def f_prime(x):
    return 3*x**2 - 6*x

x = np.linspace(-1,3,400)
y = f(x)

local_minmax_points = np.roots([3, -6, 0])
local_minmax_values = f(local_minmax_points)
inflection_point = (1, 2)

plt.figure(figsize=(10,6))
plt.plot(x, y, label = 'f(x) = x^3 - 3x^2 + 4')
plt.scatter(local_minmax_points, local_minmax_values, color='red') # Plot critical points
plt.scatter(*inflection_point, color='blue')
plt.title('Graph of f(x) = x^3 - 3x^2 + 4')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
