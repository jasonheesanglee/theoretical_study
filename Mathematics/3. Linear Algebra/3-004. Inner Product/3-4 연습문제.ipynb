# 3-4 연습문제
$a=(\sqrt3, 1), b=(1, \sqrt3), c=(-1,\sqrt3)$라고 할 때, $<a,b>, <b,c>, <c,a>$를 각각 두가지 정의식으로 계산하시오.<br>
단, 두 벡터가 이루는 각은 다음 그림과 같습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the points
a = np.array([np.sqrt(3), 1])
b = np.array([1, np.sqrt(3)])
c = np.array([-1, np.sqrt(3)])

# Plotting
plt.figure(figsize=(8,8))
plt.plot([0, a[0]], [0, a[1]], label='Oa (length = 2)')
plt.plot([0, b[0]], [0, b[1]], label='Ob (same length as Oa)')
plt.plot([0, c[0]], [0, c[1]], label='Oc (same length as Oa)')

# Adding points
plt.scatter(*a, color='red')
plt.scatter(*b, color='green')
plt.scatter(*c, color='blue')
plt.text(*a, ' a', verticalalignment='bottom', horizontalalignment='right')
plt.text(*b, ' b', verticalalignment='bottom', horizontalalignment='left')
plt.text(*c, ' c', verticalalignment='bottom', horizontalalignment='right')

# Setting the axes properties
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlim(-2, 2)
plt.ylim(-1, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Vector Plot')
plt.legend()

# Show plot
plt.show()
```
