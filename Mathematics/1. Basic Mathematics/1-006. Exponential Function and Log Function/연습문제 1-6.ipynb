# 연습문제 1-6
1. 다음 함수의 그래프를 그리시오. <br>
1-1. $y = 3^x$ <br>
1-2. $y = log_{\frac{1}{2}}x$ <br><br>
2. 다음 문제를 계산하시오. <br>
2-1. $log_{3}\sqrt{27}$ <br>
2-2. $log_{3}\frac{3}{4} + 4log_{3}\sqrt{2}$

# My Answer

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [7, 4]
plt.rcParams['figure.autolayout'] = True

x = np.linspace(0, 2, 1000)
y = 3 ** x

# plt.yscale('log')
plt.plot(x, y, c='red', lw=3, label='y = 3^x')
plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [7, 4]
plt.rcParams['figure.autolayout'] = True

x = np.linspace(0.01, 10, 1000)
y = np.log(x)/np.log(0.5)

# plt.yscale('log')
plt.plot(x, y, c='red', lw=3, label='y = log_{1/2}(x)')
plt.legend()
plt.show()
```

2-1. $log_{3}\sqrt{27}$ <br>
$=$ $log_{3}\sqrt{3^3}$ <br>
$=$ $log_{3}{3^{\frac{3}{2}}}$ <br>
$=$ $\frac{3}{2}$ <br><br>


2-2. $log_{3}\frac{3}{4} + 4log_{3}\sqrt{2}$ <br> 
$=$ $log_{3}3 - log_{3}4 + 4log_{3}\sqrt{2}$ <br> 
$=$ $1 - log_{3}{2^2} + log_{3}\sqrt{2}^4$ <br> 
$=$ $1 - log_{3}{2^2} + log_{3}4$ <br> 
$=$ $1 - 0$ <br> 
$=$ $1$ <br>
